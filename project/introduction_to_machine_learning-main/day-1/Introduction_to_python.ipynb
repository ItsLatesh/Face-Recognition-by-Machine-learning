            <u>A tuple is a collection which is ordered and unchangeable. </u><br>
            # Example <br>

            # Access touple element <br>
            print(a[0],a[1]) <br>

            2. index, find out the index of the element | Example: a.index("banana")
        </li> 
        <li>
            <strong> Set: </strong> used to store multiple items, unlike tuple it is both unordered and unindexed. Set items are unordered, unchangeable, and do not allow duplicate values. <br>
            <u>Unordered means that the items in a set do not have a defined order.</u> <br>
            <u>Sets are unchangeable, meaning that we cannot change the items after the set has been created.</u> <br>
            <u>Sets cannot have two items with the same value.</u> <br>
            # Example <br>
            a = {"apple", "banana", "cherry"} <br>
            print(a) <br>
            # Duplicate values will be ignored <br>
            a = {"apple", "banana", "cherry", "apple"} <br>
            print(a) # {"apple", "banana", "cherry"} <br>
            # Add element in the set <br>
            a.add("orange") <br>
            # Remove item from the set <br>
            a.remove("banana") <br>
            # union of two sets <br>
            b = {1,2,3} <br>
            a.union(b) <br>
            # Methods: <br>
            1. pop # remove element from the set, Example: a.remove("banana") <br>
            2. difference # to take the difference of two sets <br>
            x = {"apple", "banana", "cherry"} <br>
            y = {"google", "microsoft", "apple"} <br>
            z = x.difference(y) <br>
            print(z) # {'cherry', 'banana'}
        </li> 
        <li>
            <strong> Dictionary: </strong> Stores the data in key and value pairs <br>
            <u>Dictionary items are ordered, changeable, and does not allow duplicates. </u><br>
            # Example <br>
            a = {
                  "brand": "Ford",
                  "model": "Mustang",
                  "year": 1964
                } <br>
            # Access dictionary element <br>
            print(a["model"]) <br>
            # Change the element value <br>
            a["year"] = 2018 <br>
            # Add item in the dictionary <br>
            a["color"] = "red" <br>
            # remove item from the dictionary <br>
            a.pop("model") <br>
            del a["model"] <br>
            # copy dictionary, this is same as list <br>
            b = a <br>
            # Methods: <br>
            1. keys, return all keys of the dictionay | Example: a.keys() <br>
            2. values, return all values of the dictionary | Example: a.values() 
        </li> 
        <li>
            <strong> Bool: </strong> it is represented by the variable "bool" and it can have only two values (true or False)
        </li>  
    </ul>
