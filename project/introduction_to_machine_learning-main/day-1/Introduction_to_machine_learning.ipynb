```python
import numpy as np
```

```python
# Define the scalar
a = 5
b = 5.5
c = "avinash singh"

print("a=",a,"data type:",type(a))
print("b=",b,"data type:",type(b))
print("c=",c,"data type:",type(c))
print("---------------------------------------")

# Define the vector
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
print(v1,v2)
print("Addition of two vectors:")
v=v1+v2
print(v)
print("Vector Substraction   : ")
v=v2-v1
print(v)
print("Vector Multiplication : ")
v=v1*v2
print(v)
print("Vector Division       : ")
v=v2/v1
print(v)
print("Dot Product   : ")
# (1*5+2*6+3*7+4*8)
v=v1.dot(v2)
print(v)
print("Norm of a vector:")
norm=np.linalg.norm(v1)
print(norm)
norm=np.sqrt(np.sum(np.square(v1)))
print(norm)
print("---------------------------------------")
```

```python
# Define matrix
m = np.matrix('1,2,3; 4,5,6; 7,8,9')
print(m)
transpose_m = m.T
print("matrix Transpose:")
print(transpose_m)
print("Determinant  of a matrx:")
determinant_m = np.linalg.det(m)
print(determinant_m)
print("Inverse of a matrix")
inverse_m = np.linalg.inv(m)
print(inverse_m)
```

```python
# Euclidean Distance
# Define the vector
v1 = np.array([3,4])
v2 = np.array([7,8])
# find the distance
d1 = np.linalg.norm(v1-v2)
d2 = np.sqrt(np.sum(np.square(v1-v2)))
print("Euclidean distance:",d1)
print("Euclidean distance:",d2)
# Manhattan distance
d1=np.linalg.norm(v1-v2,ord=1)
d2=np.sum(np.abs(v1-v2))
print("Manhattan distance:",d1)
print("Manhattan distance:",d2)
# Chebyshev distance
d1=np.linalg.norm(v1-v2,ord=np.inf)
d2=np.abs(v1-v2).max()
print("Chebyshev distance:",d1)
print("Chebyshev distance:",d2)
# Cosine distance
d=np.dot(v1,v2)/(np.linalg.norm(v1)*(np.linalg.norm(v2)))
print("Cosine distance:",d)
```

```python
x=np.array([2,30,8,11,21,16,5,25])
# Calculating mean
m1=np.mean(x)
m2=np.sum(x)/len(x)
print("Mean:",m1)
print("Mean:",m2)

# Calculating Median
x1=np.sort(x)
m1=np.median(x)
index=len(x1)//2
if len(x1)%2==0:
    m2=(x1[index-1]+x1[index])/2
else:
    m2=x1[index]
print("Median:",m1)
print("Median:",m2)

# Calculating mode
x=np.array([2,3,5,5,7,9,9,9,10,12])
vals,counts = np.unique(x, return_counts=True)
print(x,vals,counts)
index = np.argmax(counts)
print("Mode:",vals[index])
```

```python
age = np.array([14,15,16,17,14,18,20])
height = np.array([158,162,165,167,155,170,173])
weight = np.array([80,60,65,55,58,65,75])

# Lets plot the data
import matplotlib.pyplot as plt
N = len(age)
x = age
y = height
colors = np.random.rand(N)
point_size = 100  
plt.scatter(x, y, s=point_size, c=colors, alpha=0.5)
plt.title('Age VS Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()

# Age and Weight distribution
x = age
y = weight
colors = np.random.rand(N)
point_size = 100  
plt.scatter(x, y, s=point_size, c=colors, alpha=0.5)
plt.title('Age VS Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```

```python
# standard deviation
standard_deviation_age = np.std(age)
print("Standard Deviation -Age:",standard_deviation_age)
standard_deviation_height = np.std(height)
print("Standard Deviation -Height:",standard_deviation_height)
standard_deviation_weight = np.std(weight)
print("Standard Deviation -Weight:",standard_deviation_weight)

# Variance
variance_age = np.var(age)
print("Variance -Age:",variance_age)
variance_height = np.var(height)
print("Variance -Height:",variance_height)
variance_weight = np.var(weight)
print("Variance -Weight:",variance_weight)

# Co-Variance between age and height
covariance_age_height = np.cov(age,height)
print("Covariance  -Age and Height:",covariance_age_height)

# Co-Variance between age and weight
covariance_age_weight = np.cov(age,weight)
print("Covariance -Age and Weight:",covariance_age_weight)

# Co-relation between age and height
corelation_age_height = np.corrcoef(age,height)
print("Corelation  -Age and Height:",corelation_age_height)

# Co-Variance between age and weight
corelation_age_weight = np.corrcoef(age,weight)
print("Corelation -Age and Weight:",corelation_age_weight)
```
